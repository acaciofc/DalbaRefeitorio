```python
#1. Faça um programa que leia os ips

ip = open('ip.txt', 'r')
lista = ip.readlines()
valido = open('valido.txt', 'w')
invalido = open('invalido.txt', 'w')
valido.write('*** IPs Válidos ***\n')
invalido.write('*** IPs Inválidos ***\n')
for i in range(0, len(lista)):
    aux = lista[i].split('.')
    for x in range(0, len(aux)):
        if int(aux[x]) < 255:
            if x == 3:
                valido.write(lista[i])
        else:
            invalido.write(lista[i])
            break
```

```python
#2. Quando falamos sobre erros e exceções
#a) Erro de sintaxe
#Os erros de sintaxe são geralmente causados por erros na estrutura do código, como falta de parênteses, 
#pontos de vírgula ausentes, indentação incorreta, entre outros. 

#b) Erro de tempo de execução
# Erros de lógica. Alguns dos erros de tempo de execução mais comuns incluem exceções, como 
# IndexError (índice fora do intervalo), ValueError (valor inválido), TypeError (tipo inválido), entre outros.

#c) Erro semântico
# Ocorrem quando o programa não apresenta erro de sintaxe e não ocorre um erro durante a execução. 
# O programa é executado do começo ao fim, mas ele não produz o resultado desejado.
# Ex. Uma função chamada soma que retorna uma subtração
```

```python
def operacao():
    print('+ -> SOMA')
    print('- -> SUBTRAÇÃO')
    print('/ -> DIVISÃO')
    print('* -> MULTIPLICAÇÃO')

while True:
    try:
        primeiro = int(input(print('Digite o primeiro numero: ')))
    except:
        print('Entrada inválida')
    else:
        break
while True:
    try:
        segundo = int(input(print('Digite o segundo numero: ')))
    except:
        print('Entrada inválida')
    else:
        break

operacao()
operador = input(print('QUAL OPERAÇÃO DESEJA?'))
        
if operador not in ['+', '-', '/', '*']:
    while operador not in ['+', '-', '/', '*']:
        mensagem = ('%s <-> NÃO É UM OPERADOR VÁLIDO' %operador)
        print(len(mensagem) * '*')
        print(mensagem)
        print(len(mensagem) * '*')
        operacao()
        operador = input(print('QUAL OPERAÇÃO DESEJA?'))

import operator
calculo = { "+": operator.add, "-": operator.sub,"*": operator.mul,"/":operator.truediv}
print('A OPERAÇÃO DE ' + str(primeiro) + ' '+ operador + ' ' + str(segundo) + ' = ', end='')
print (calculo[operador](primeiro,segundo))

```
